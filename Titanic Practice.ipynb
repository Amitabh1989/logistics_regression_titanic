```python
def impute_age(cols, m):
    Age = cols[0]
    Pclass = cols[1]
    m1 = m[0]
    m1 = m[1]
    m1 = m[2]
    
    if pd.isnull(Age):
        print("P class is : %s" % Pclass)
        if int(Pclass) == 1:
            print("Returning 1")
            return m1
        elif int(Pclass) == 2:
            print("Returning 2")
            return m2
        elif Pclass == 3.0:
            print("Returning 3")
            return m3
    else:
        print("Returning Original")
        return Age
    print("\n\n")
```

```python
m1 = df[(df['Pclass'] == 1.0)]["Age"].mean()
m2 = df[(df['Pclass'] == 2.0)]["Age"].mean()
m3 = df[(df['Pclass'] == 3.0)]["Age"].mean()
m = [m1, m2, m3]

print("M1 : %s" %m1)
print("M2 : %s" %m2)
print("M3 : %s" %m3)

df["Age"] = df[['Age', 'Pclass']].apply(impute_age, args=(m,), axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop("Survived", axis=1)
X.head()
```

```python
y = df["Survived"]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
log_r = LogisticRegression(max_iter=300)
```

```python
log_r.fit(X_train, y_train)
```

```python
y_pred = log_r.predict(X_test)
```

```python
sns.distplot((y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
